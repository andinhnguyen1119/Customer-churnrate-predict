# 4. Data Preprocessing

```python
df2 = df.copy()
```

Kiểm tra các giá trị trùng lặp:

```python
df2[df2.duplicated()]
```

Chuyển các giá trị không hợp lệ (giá trị âm) về giá trị null:

```python
df2.loc[df2['days_since_last_login'] < 0, 'days_since_last_login'] = np.nan
df2.loc[df2['avg_time_spent'] < 0, 'avg_time_spent'] = np.nan
df2.loc[df2['points_in_wallet'] < 0, 'points_in_wallet'] = np.nan
```

Kiểm tra thuộc tính nào bị thiếu giá trị, nan, các thuộc tính có giá trị không hợp lệ:

```python
def display_invalid(df, feature_cols):
    df = df.replace('?', None)
    n_rows = df.shape[0]
    for col in feature_cols:
        null_count = df[col].isnull().sum()
        na_count = df[col].isna().sum()
        error_count = (df[col] == 'Error').sum()
        unknown_count = (df[col] == 'Unknown').sum()
        if null_count > 0:
            print(f"{col} has {null_count*100/n_rows:.2f}% null values.")
        elif na_count > 0:
            print(f"{col} has {na_count*100/n_rows:.2f}% na values.")
        elif error_count > 0:
            print(f"{col} has {error_count*100/n_rows:.2f}% error values.")
        elif unknown_count > 0:
            print(f"{col} has {unknown_count*100/n_rows:.2f}% unknown values.")
            
display_invalid(df2, feature_cols)
```

Điền các dữ liệu bị khuyết, NaN, các thuộc tính có giá trị không hợp lệ:

- Xóa các cột `customer_id`, `Name`, `security_no`, `referral_id`, `last_visit_time` để tăng hiệu suất trong quá trình training model.
- Cột `gender` có một số dòng có giá trị là Unknow -> chuyển qua 'F'
- Cột `region_category` có một số dòng có giá trị là null -> chuyển qua 'Town'
- Cột `joined_through_referral` có một số dòng có giá trị là ? -> chuyển thành 'No'
- Cột `preferred_offer_types` có một số dòng có giá trị là null -> chuyển thành 'Gift Vouchers/Coupons'
- Cột `medium_of_operation` có một số dòng có giá trị là ? -> chuyển thành 'Desktop'
- Cột `avg_frequency_login_days` có một số dòng có dữ liệu là error -> chuyển thành 0
- Cột `points_in_wallet` có một số dòng có dữ liệu là null -> chuyển thành median của cột dữ liệu đó

```python
data_drop = df2[df2['churn_risk_score']==-1].index
df2.drop(data_drop, inplace=True)

df2['gender'] = df2['gender'].apply(lambda x: 'F' if x == 'Unknown' else x)
df2['joined_through_referral'] = df2['joined_through_referral'].apply(lambda x: 'No' if x == '?' else x)
df2['medium_of_operation'] = df2['medium_of_operation'].apply(lambda x: 'Desktop' if x == '?' else x)
df2['region_category'].fillna('Town', inplace=True)
df2['preferred_offer_types'].fillna('Gift Vouchers/Coupons', inplace=True)
# Chuyển error -> NaN sau đó xử lý 1 lần
df2['avg_frequency_login_days'] = df2['avg_frequency_login_days'].apply(lambda x: np.nan if x == 'Error' else x)
df2['avg_frequency_login_days'] = pd.to_numeric(df2['avg_frequency_login_days'])
df2.loc[df2['avg_frequency_login_days'] < 0, 'avg_frequency_login_days'] = np.nan
df2['avg_frequency_login_days'].fillna(df2['avg_frequency_login_days'].median(), inplace=True)
df2['avg_time_spent'].fillna(df2['avg_time_spent'].median(), inplace=True)
df2['points_in_wallet'].fillna(df2['points_in_wallet'].median(), inplace=True)
df2['days_since_last_login'].fillna(df2['days_since_last_login'].median(), inplace=True)
```

Xóa những cột có độ đa dạng cao và đóng góp thấp:

```python
# Bỏ các cột customer_id, Name, security_no, referral_id, last_visit_time vì có độ đa dạng giá trị cao.
# Vì complaint_status đã thể hiện ý nghĩa của past_complaint nên bỏ cột past_complaint.
# Bỏ thêm preferred_offer_types, region_category, joined_through_referral, days_since_last_login, avg_time_spent, used_special_discount, offer_application_preference, complaint_status, internet_option, medium_of_operation, gender, age, days_since_joined, gender, age vì độ tương quan với churn_risk_score thấp và độ tương quan với các thuộc tính liên quan thấp.

colsToRemove = ['preferred_offer_types', 'region_category', 'joined_through_referral', 'days_since_last_login', 'avg_time_spent', 'used_special_discount','customer_id', 'offer_application_preference',
                'Name', 'security_no', 'referral_id', 'last_visit_time', 'past_complaint', 'complaint_status', 'internet_option', 'medium_of_operation', 'gender', 'age', 'days_since_joined']
df2.drop(colsToRemove, inplace=True, axis=1)
df2.head()
```

Encode Categorical data:

```python
from sklearn.preprocessing import LabelEncoder

def apply_feature_engineering(df):
    """
    Apply all feature engineering to transform your data into number
    :param df: pandas DataFrame
    :return: pandas DataFrame
    """
    df['churn_risk_score'] = df['churn_risk_score'].astype(int)
    df["membership_category"] = df["membership_category"].apply(lambda x: 0 if x=="No Membership" 
                                                                     else 1 if x=="Basic Membership" 
                                                                     else 2 if x=="Silver Membership"
                                                                     else 3 if x=="Gold Membership"
                                                                     else 4 if x=="Premium Membership"
                                                                     else 5)

    LE_cols = ['feedback']
    LE = LabelEncoder()
    for i in LE_cols:
        df[i] = df[[i]].apply(LE.fit_transform)

    
    return df

data = apply_feature_engineering(df2)
display(data.describe())
display(data.head())
display(data.corr())
sns.heatmap(data.corr()) # Show heatmap after normalized data
```

Split Train-test:

```python
def prepare_X_y(df):
    """
    Feature engineering and create X and y
    :param df: pandas dataframe
    :return: (X, y) output feature matrix (dataframe), target (series)
    """
    feature_names = df.columns.tolist()                                
    feature_names.remove("churn_risk_score")

    X = df[feature_names].values
    y = df.churn_risk_score.values
    return X, y

X, y = prepare_X_y(df2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

def split_train_test(X, y, train_size=0.75):
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=train_size, random_state=2023)
    print('Training: ' + str(X_train.shape))
    print('Test: ' + str(X_test.shape))

    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = split_train_test(X, y)
```

# 6. Apply machine learning model

Upsampling:

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_train_oversampling, y_train_oversampling = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.preprocessing import RobustScaler
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

pipe = Pipeline(steps=[
    ("scaler", RobustScaler(quantile_range=(5, 95))),
    ("svc", SVC(kernel='rbf', decision_function_shape='ovo', cache_size=300))
])

param_grid = {
    'svc__C': [80, 100, 200, 400, 600],
    'svc__gamma': [2, 4, 6, 8, 10, 12]
}

search = GridSearchCV(pipe, param_grid, cv=5, n_jobs=5, verbose=2)
search.fit(X_train_oversampling, y_train_oversampling)
print("Best parameter (CV score=%0.3f):" % search.best_score_)
print(search.best_params_)

from sklearn.metrics import classification_report, confusion_matrix
y_pred = search.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# Vẽ confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test , y_pred)
plt.imshow(cm, cmap='flare')
plt.colorbar()
# Đặt tên cho trục x và trục y
tick_marks = np.arange(5)
plt.xticks(tick_marks, ['Class 1', 'Class 2', 'Class 3', 'Class 4', 'Class 5'])
plt.yticks(tick_marks, ['Class 1', 'Class 2', 'Class 3', 'Class 4', 'Class 5'])
# Hiển thị giá trị của từng ô trong confusion matrix
thresh = cm.max() / 2.
for i, j in np.ndindex(cm.shape):
    plt.text(j, i, format(cm[i, j], 'd'),
             ha="center", va="center",
             color="white" if cm[i, j] > thresh else "black")

# Đặt tên cho trục x và trục y
plt.xlabel('Predicted Class')
plt.ylabel('True Class')
plt.tight_layout()
plt.show()
```
